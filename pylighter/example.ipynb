```python
from numpy import linspace
import pylighter as pl

%matplotlib inline
```

```python
# Only required arguments are tokens and scores.
tokens = 'Color is more important than other aspects of figure style because color can reveal patterns in the data if used effectively or hide those patterns if used poorly. There are a number of great resources to learn about good techniques for using color in visualizations, I am partial to this series of blog posts from Rob Simmon and this more technical paper. The matplotlib docs also now have a nice tutorial that illustrates some of the perceptual properties of the built in colormaps.'.split()
scores = linspace(0, 1, len(tokens))
```

```python
# Default rendering.
fig, ax = pl.render(tokens, scores)
```

```python
# Changle number of typical words per line.
fig, ax = pl.render(tokens, scores, typical_wpl=5)
```

```python
# Change the color palette.
fig, ax = pl.render(tokens, scores, palette='Paired')
```

```python
# Change the font size and family.
fig, ax = pl.render(tokens, scores, fontsize=30, name='fantasy')
```

```python
# Change the alpha value and offset factor.
fig, ax = pl.render(tokens, scores, alpha=0.3, offset_factor=2.)
```
